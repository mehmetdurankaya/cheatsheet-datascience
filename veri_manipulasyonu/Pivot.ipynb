# PIVOT

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.groupby("sex")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
# pivot ile table
```

```python
titanic.pivot_table("survived",index="sex",columns="class")
```
