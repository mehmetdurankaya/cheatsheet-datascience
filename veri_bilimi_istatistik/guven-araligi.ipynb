# Güven Aralığı

### Anakütle parametresinin tahmini değerini kapsayabilecek iki sayıdan oluşan bir aralık bulunması işlemidir.
* Ölçümün hassasiyetinin bir göstergesidir.
* Yapılan tahminlerin nekadar güvenilir olduğu ile ilgili bize değer sunar.

* ÖRNEK
* Web sitesinde geçirilen sürenin güven aralığı nedir?

    - Ortalama: 180 saniye
    - Standart sapma: 40 saniye
      


* istatistiksel olarak %95 güvenilirlik ile web sitemizde geçirilen ortalam süre 172 ile 188 saniye arasındadır.
* Bu aralık aslında bize şunu söyler: Gerçekleştirilecek olan 100 denemeden 95' inde ortalama bu aralıktadır.
