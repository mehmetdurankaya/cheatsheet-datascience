# Veri Bilimi için İstatistik
## Örnek Teorisi

* Örneklem
* Örneklem Dağılımı
* Merkezi Limit Teorimi

## İSTATİSTİK veri biliminin "bilim" makine öğrenmesinin "öğrenme" kısmıdır.

* Popülasyon(Ana kitle)
* Örneklem: Ana kitleyi popülasyonu temsil ettiği düşünülen alt kümedir. İkiye ayrılır
   - Olasılıksal
   - Olasılıksal olmayan
* Örneklem dağılımı: Popülasyondan birden fazla örneklem çekildiğinde ve bunların dağılımı ile ilgilenelismesidir.

* Merkezi Limit Teoremi: Bağımsız veya aynı dağılıma sahip rassal değişkenlerin toplamı ya da aritmetik ortalaması yaklaşık olarak normal dağılmaktadır.
   - Bir popülasyona odaklanıp onun üzerinden çıkarım yapmak üzere örneklemlerle ilgilenildiğimizde bu konuya eğilmiş oluruz
