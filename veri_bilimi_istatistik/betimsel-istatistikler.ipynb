<span style="font-size: 12px;">&#x25A0;</span> <span style="font-size: 24px; margin-left: 10px;margin-right: 200px;"><b>Kovaryans</b></span><br>
#### iki değişkenin arasındaki ilişkinin değişkenlik ölçüsüdür.
#### varyans ortalama etrafındaki dağılımın ölçüsü
#### standartsapma ile beraber düşünüldüğünde standart sapma karesi alındığında varyans elde elidiyordu
#### kovaryans ise iki değişken arasındaki ilişkinin değişkenlik ölçüsüdür.
#### iki rasgele değişkenin kendi ortalamalarından olan sapmalarının çarpımlarının beklenen değeridir. Böylece iki değişkenin birlikte ortaya çıkardığı değişim incelenmiş olur

<span style="font-size: 12px;">&#x25A0;</span> <span style="font-size: 24px; margin-left: 10px;margin-right: 200px;"><b>Korelasyon</b></span><br> 
#### İki değişken arasındaki ilişkiyi, ilişkinin anlamlı olup olmadığını, ilişkinin şiddetini ve yönünü ifade eden istatistiksel bir tekniktir.

### Korelasyon ve Kovaryans arasındaki fark
* Kovaryans iki değişken arasında ilişki olabilir bunun değişkenliği nedir ile ilgilenir
* Korelasyon iki değişken arasındaki ilişkiyi anlamlı olup olmadığını yönü ve şiddeti ile ilgilenir.
