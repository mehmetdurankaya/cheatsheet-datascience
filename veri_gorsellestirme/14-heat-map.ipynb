# Heat Map (Isı Haritası)

* Uzun vadeli verilerde yani içerisinde zaman serisi verileri yıl ve ay gibi değişkenler olduğunda bu dönemlere karşılık gözlemlenecek olan bazı sayısal değişken değerleri olduğunda yada bunun dışında daha büyük ölçekli belirli periyotlarla tekrar eden olayları görmek istediğimizde bize bilgiler sunan bir grafiktir.
* Sadece zamansal bağlamda değil eğer elimizde çok sınıflı kategorik değişken ve bunu belirli bir sayısal değişken açısından görselleştirmek gerektiğinde çok işe yarar

```python
import seaborn as sns
flights = sns.load_dataset('flights')
df=flights.copy()
df.head()
```

```python
df.shape
```

```python
df["passengers"].describe()
```

```python
df = df.pivot(index="month", columns="year", values="passengers")
```

```python
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df,annot=True,fmt="d");
```

```python
sns.heatmap(df,annot=True,fmt="d");
```

```python
sns.heatmap(df,annot=True,fmt="d",cbar=False);
```
