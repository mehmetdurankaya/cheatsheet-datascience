# Violin GrafiÄŸi

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.catplot(y = "total_bill", kind = "violin", data=df);
```

```python
sns.catplot(x="day" , y = "total_bill", kind = "violin", data=df);
```
