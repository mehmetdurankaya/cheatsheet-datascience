# Histogram ve Yoğunluk

```python
import seaborn as sns
sns.set_theme()
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
# price değişkenine odaklanıyoruz
```

```python
sns.histplot(df["price"],bins=10, kde = False); # kde yoğunluk fonksiyonunda grafiğe işlenip işlenmeyeceği ile ilgili ayarlama
```
