# Korelasyon Grafikleri

### Korelasyon değişkenler arasındaki ilişkiyi ifade eden istatistiksel bir terimdir.
* Değişkenler arasındaki ilişki incelendiğinde buna korelasyon denir.
* İki değişken arasındaki ilişkiyi ifade etmek için kullanılan ve ençok bilinen yaklaşım Scatterplot(yaklaşım grafiği) bize sayısal değişkenler arasındaki ilişkiyi gösterir yani bu iki değişken bir arada değerlendirildiğinde veri setinin içerisde ortaya çıkan yapı hakkında bize bilgi verir.

## Scatterplot

* total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
* tip: bahşiş
* sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
* smoker: grupta sigara içen var mı?
* day: gün(3=Thur,t=Fri,5=Sat,6=Sun)
* time: ne zaman? (0=Day,1=Night)
* size: grupta kaç kişi var?

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x="total_bill", y= "tip",data = df);
```

# Çaprazlamalar

* Bir değişkende belirli bir yönde bir artış olduğunda diğer bir değişkende de bir artış gözlemleniyor ancak bu kırılımda bu gözlenmemiş gibi farklılaştırma yapmaya yarar 

```python
sns.scatterplot(x="total_bill", y= "tip",hue="time",data = df);
```

```python
sns.scatterplot(x="total_bill", y= "tip",hue="time",style="time",data = df);
```

```python
sns.scatterplot(x="total_bill", y= "tip",hue="day",style="time",data = df);
```

```python
sns.scatterplot(x="total_bill", y= "tip",size="size",data = df);
```

```python
sns.scatterplot(x="total_bill", y= "tip",hue="size",size="size",data = df);
```
