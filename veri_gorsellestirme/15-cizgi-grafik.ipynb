* Zamana bağlı değişkenlerin olduğu veri setlerinde ve nesnelerin interneti gibi senaryolarda makinaların ürettiği verileri görselleştirmek için kullanabileceğimiz bir veri görselleştirme tekniğidir.

```python
import seaborn as sns
fmri= sns.load_dataset("fmri")
df= fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
df.groupby("signal").count()
```

```python
df.groupby("timepoint")["signal"].describe().T
```

```python
sns.lineplot(x = "timepoint",y="signal",data=df);
```

```python
sns.lineplot(x = "timepoint",y="signal",hue="event",data=df);
```

```python
sns.lineplot(x = "timepoint",y="signal",hue="event",style="event",data=df);
```

```python
sns.lineplot(x = "timepoint",
             y="signal",
             hue="event",
             style="event",
             markers=True,
             dashes=False,
             data=df);
```

```python
sns.lineplot(x = "timepoint",
             y="signal",
             hue="region",
             style="event",
             data=df);
```
